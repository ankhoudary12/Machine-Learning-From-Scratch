```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%pylab inline
from copy import deepcopy
#necessary imports 
```

```python
link = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# let's use the popular iris data set
df=pd.read_csv(link, header=None)
```

```python
df.columns=['slength','swidth','plength','pwidth','label']
#get correcxt feature labels 
```

```python
x=df.drop('label',axis=1)
x=np.array(x)
```

```python
class K_Means():
    
    def __init__(self, k=3):
        
        self.k=k
        self.margin=0.0001 #from scikit learn
        self.inertia=[] #keep record of learning 
        
    def __distance(self, x1,x2, axis):
        
        '''
        distance helper funtion. Don't need square root as we only care about comparison not absolute values
        '''
        
        return np.sum((x1-x2)**2,axis=axis)
    
    def fit(self, data):
        
        data=np.array(data) #make sure data is in a numpy array
        
        centroids=data[np.random.choice(len(x),self.k, replace=False)] #randomly assign k points to be centroids
        
        margin=50 #set arbitrary margin for looping
        
        while margin > self.margin: #loop until difference between updated centroids is less than acceptable margin 
            
            #find distance from each point to each centroid
            
            distances=np.array([self.__distance(data,y,axis=1) for y in centroids]).T 
            
            labels=np.argmin(distances, axis=1) #find the nearest centroid for each point
            
            #find the total distance from each point to its assigned centroid
            
            inertia_score= sum([self.__distance(feat,centroids[labels[i]],0) for i, feat in enumerate(data)])
            
            self.inertia.append(inertia_score)
            
            old_centroids=deepcopy(centroids) #save the old centroids to compare when they are updated
            
            for i in range(self.k):
                points=[feat for index, feat in enumerate(data) if labels[index]==i] #get all points for each centroid
                centroids[i]=np.mean(points, axis=0) #make new centroid the average of all its points
                
            margin=sum(self.__distance(old_centroids,centroids,0)) #find distance between old and new centroids
            
            #if the difference is less than the margin we allowed, the loop breaks as no improvement is being made
            #on updating the centroids any further
            
        self.centroids=centroids
        
        self.labels=labels
        
    def centroids(self):
        return self.centroids
    
    def inertia(self):
        return self.inertia
    
    def labels(self):
        return self.labels
    
    def SSE(self):
        return self.inertia[-1]
            
            
        
        
        
        
        
```

```python
km=K_Means(3)
```

```python
km.fit(x)
```

```python
df['cluster']=km.labels
```

```python
plt.figure(figsize=(6,6)) #let's see how our clusters compare to the original dataset
sns.lmplot(data=df ,x='slength', y='plength',hue='label',
                   fit_reg=False, legend=True, legend_out=False)

sns.lmplot(data=df ,x='slength', y='plength',hue='cluster',
                   fit_reg=False, legend=True, legend_out=False)
```

Looks like we did a pretty good job. Our clusters resemble the original three clusters. Yet, we knew from the start how many clusters there was supposed to be. What if we didn't know. How could we deduce the correct amount? 

To do this, we can look at the final inertia (or SSE) of our clusters. This is the total distance for all points to their corresponding clusters. As we increase the number of k or clusters, we should expect this number to go down. However, there should be a point where this value elbows, or no longer offers a lower SSE as you increase k. A simple graph will help illustrate this point.

```python
inertia_scores=[]
for i in range(2,15):
    km=K_Means(i)
    km.fit(x)
    inertia_scores.append(km.SSE())
    


plt.figure(figsize(20,10))
plt.title('SSE for varying clusters',size=20)
plt.xlabel('Number of clusters',size=15)
plt.ylabel('SSE',size=15)
plt.plot(np.arange(2,15), inertia_scores)
```

From Looking at the Graph, it looks like 6 clusters might be the ideal amount of clusters. However, we know from the data set that we are only supposed to have 3!!! This points out a glaring weakness of the k_means method. By not knowing the appropriate amount of clusters beforehand we may get incorrect results. So, using the SSE graph is a good place to start but by no means a perfect method.

Now, let's see how the sklearn counterpart compares

```python
from sklearn.cluster import KMeans
```

```python
km_sklearn=KMeans(n_clusters=3)
```

```python
km_sklearn.fit(x)
```

```python
df['cluster_sklearn']=km_sklearn.labels_
```

```python
plt.figure(figsize=(6,6)) #let's see how our clusters compare to the original dataset
sns.lmplot(data=df ,x='slength', y='plength',hue='label',
                   fit_reg=False, legend=True, legend_out=False)

sns.lmplot(data=df ,x='slength', y='plength',hue='cluster_sklearn',
                   fit_reg=False, legend=True, legend_out=False)
```

The sklearn algroithm produced very similar results to our custom K Means!!!
